### Question 2

On nous demande maintenant de tracer sur un graphe quelques valeurs de l'erreur d'interpolation E n (f) en fonction de n et de commenter le résultat. Pour cela, nous pouvons utiliser la formule de l'erreur d'interpolation donnée dans l'énoncé, ainsi que la fonction "max" pour calculer la valeur maximale de l'erreur sur l'intervalle [-5, 5].

Voici le code pour calculer l'erreur d'interpolation pour n allant de 1 à 20 et la tracer sur un graphique :

```matlab

a = -5;
b = 5;

n_values = 1:20;

On peut observer que l'erreur diminue rapidement avec n jusqu'à n = 5, puis commence à augmenter de manière significative pour n > 5. Ce phénomène est connu sous le nom de "phénomène de Runge" et est dû à l'interpolation d'une fonction oscillante avec des polynômes de degré élevé. En effet, plus le degré du polynôme d'interpolation est élevé, plus il peut osciller entre les points d'interpolation, ce qui peut conduire à une mauvaise approximation de la fonction réelle. Pour éviter ce phénomène, il est souvent recommandé d'utiliser des méthodes d'interpolation non polynomiales, telles que l'interpolation par splines.

On nous demande maintenant de tracer sur un graphe quelques valeurs de l'erreur d'interpolation E n (f) en fonction de n et de commenter le résultat. Pour cela, nous pouvons utiliser la formule de l'erreur d'interpolation donnée dans l'énoncé, ainsi que la fonction "max" pour calculer la valeur maximale de l'erreur sur l'intervalle [-5, 5].

Voici le code pour calculer l'erreur d'interpolation pour n allant de 1 à 20 et la tracer sur un graphique :

```matlab
% Définition de la fonction f(x) = 1/(1+x^2) sur l'intervalle [-5, 5]
f = @(x) 1./(1+x.^2);

% Définition de l'intervalle [-5, 5]
a = -5;
b = 5;

% Calcul de l'erreur d'interpolation pour n allant de 1 à 20
n_values = 1:20;
E_n = zeros(size(n_values));
for i = 1:length(n_values)
    n = n_values(i);
    % Calcul des n+1 points équidistants dans l'intervalle [a, b]
    xp = linspace(a, b, n+1);
    
    % Calcul des coefficients du polynôme d'interpolation de Lagrange
    p = lagrange_interpolation(xp, f(xp));
    
    % Calcul de l'erreur d'interpolation
    x = linspace(a, b, 1000);
    y = f(x);
    E_n(i) = max(abs(y - polyval(p, x)));
end

% Tracé de l'erreur d'interpolation en fonction de n
plot(n_values, E_n, 'o-');
title('Erreur d''interpolation de f(x)');
xlabel('n');
ylabel('E_n');
```

En exécutant ce code, vous devriez obtenir un graphique montrant l'erreur d'interpolation en fonction de n. On peut observer que l'erreur diminue rapidement avec n jusqu'à n = 5, puis commence à augmenter de manière significative pour n > 5. Ce phénomène est connu sous le nom de "phénomène de Runge" et est dû à l'interpolation d'une fonction oscillante avec des polynômes de degré élevé. En effet, plus le degré du polynôme d'interpolation est élevé, plus il peut osciller entre les points d'interpolation, ce qui peut conduire à une mauvaise approximation de la fonction réelle. Pour éviter ce phénomène, il est souvent recommandé d'utiliser des méthodes d'interpolation non polynomiales, telles que l'interpolation par splines.

```matlab
% Dfinition de la fonction f(x) = 1/(1+x^2) sur l'intervalle [-5, 5]
f = @(x) 1./(1+x.^2);

% Dfinition de l'intervalle [-5, 5]
a = -5;
b = 5;

% Calcul de l'erreur d'interpolation pour n allant de 1  20
n_values = 1:20;
E_n = zeros(size(n_values));
for i = 1:length(n_values)
    n = n_values(i);
    % Calcul des n+1 points quidistants dans l'intervalle [a, b]
    xp = linspace(a, b, n+1);
    
    % Calcul des coefficients du polynme d'interpolation de Lagrange
    p = lagrange_interpolation(xp, f(xp));
    
    % Calcul de l'erreur d'interpolation
    x = linspace(a, b, 1000);
    y = f(x);
    E_n(i) = max(abs(y - polyval(p, x)));
end

% Trac de l'erreur d'interpolation en fonction de n
plot(n_values, E_n, 'o-');
title('Erreur d''interpolation de f(x)');
xlabel('n');

```

En excutant ce code, vous devriez obtenir un graphique montrant l'erreur d'interpolation en fonction de n. On peut observer que l'erreur diminue rapidement avec n jusqu' n = 5, puis commence  augmenter de manire significative pour n > 5. Ce phnomne est connu sous le nom de "phnomne de Runge" et est d  l'interpolation d'une fonction oscillante avec des polynmes de degr lev. En effet, plus le degr du polynme d'interpolation est lev, plus il peut osciller entre les points d'interpolation, ce qui peut conduire  une mauvaise approximation de la fonction relle. Pour viter ce phnomne, il est souvent recommand d'utiliser des mthodes d'interpolation non polynomiales, telles que l'interpolation par splines.
